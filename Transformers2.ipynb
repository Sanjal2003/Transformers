```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

# Define the prompt
prompt = "In a distant future, humanity has discovered"

# Encode the prompt with attention mask and padding id
input_ids = tokenizer(prompt, return_tensors="pt").input_ids
attention_mask = torch.ones_like(input_ids)  # Creates an attention mask to prevent warnings

# Generate the text
output = model.generate(
    input_ids,
    attention_mask=attention_mask,         # Use attention mask to ensure consistent behavior
    max_length=200,                        # Length of generated text
    num_return_sequences=1,
    no_repeat_ngram_size=2,
    top_k=50,
    top_p=0.95,
    temperature=0.7,                        # Lower temperature for more coherent text
    pad_token_id=tokenizer.eos_token_id,    # Set pad token to eos token to avoid warnings
    do_sample=True
)

# Decode and print the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```
